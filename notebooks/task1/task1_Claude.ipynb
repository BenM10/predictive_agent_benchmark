```python
import pandas as pd
import numpy as np
```

## 1. Data Ingestion

```python
df = pd.read_csv("../../data/processed/airline_passenger_satisfaction_full.csv")

print(f"Rows:    {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
```

## 2. Schema Inspection

```python
print("Column names:")
for col in df.columns:
    print(f"  - {col}")
```

```python
print("Data types:")
print(df.dtypes.to_string())
```

## 3. Missing Value Check

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(4)
missing_report = pd.DataFrame({
    "Missing Count": missing,
    "Missing %": missing_pct
})
cols_with_missing = missing_report[missing_report["Missing Count"] > 0]

if len(cols_with_missing) == 0:
    print("No missing values found.")
else:
    print(f"Columns with missing values ({len(cols_with_missing)}):")
    display(cols_with_missing)
    print(f"\nTotal missing cells: {missing.sum():,}")
```

## 4. Summary Statistics

```python
print("Numeric columns – summary statistics:")
display(df.describe().T)
```

```python
print("Categorical columns – value counts:")
cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    print(f"\n{col}:")
    display(df[col].value_counts().to_frame())
```

## 5. Data Preview

```python
display(df.head())
```

## 6. Save Cleaned Dataset

Only `Arrival Delay in Minutes` has missing values (393 rows, 0.30%).  
We drop those rows and save the cleaned dataset.

```python
df_clean = df.dropna(subset=["Arrival Delay in Minutes"]).reset_index(drop=True)
df_clean.to_csv("../../data/processed/Claude_clean.csv", index=False)

print(f"Original shape : {df.shape[0]:,} rows x {df.shape[1]} columns")
print(f"Clean shape    : {df_clean.shape[0]:,} rows x {df_clean.shape[1]} columns")
print(f"Rows dropped   : {df.shape[0] - df_clean.shape[0]:,}")
print("Saved to: data/processed/Claude_clean.csv")
```
