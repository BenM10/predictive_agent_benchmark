# Preparatory Dataset Setup  
Currently .csv's exist as seperate train and test sets, so will combine them first to ensure all agents start from a fair and standardised beginning. 

```python
import pandas as pd
from pathlib import Path

# Define file paths
train_path = Path("../data/raw/train.csv")
test_path = Path("../data/raw/test.csv")
output_path = Path("../data/processed/airline_passenger_satisfaction_full.csv")

# Load datasets
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)

# Combine them
df = pd.concat([train, test], ignore_index=True)

# Drop any obvious index-like columns if present
for col in ["Unnamed: 0", "Index", "index"]:
    if col in df.columns:
        df = df.drop(columns=[col])

# Save combined dataset
df.to_csv(output_path, index=False)

# Preview result
print("Combined dataset saved successfully.")
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

df.head()
```
